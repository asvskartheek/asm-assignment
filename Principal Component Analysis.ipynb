```python
import pandas as pd
```

```python
df = pd.read_csv('expectations_responses.csv')
df.head()
```

```python
dic = {}
for i in range(1,4):
    dic[i] = []
for index,row in df.iterrows():
    max_index = -1
    max_val = 0
    for i in range(1,4):
        if row[str(i)]>max_val:
            max_val = row[str(i)]
            max_index = i
    dic[max_index].append(index+1)
        
print(dic)
```

```python
x = df.values
from sklearn.decomposition import PCA
pca = PCA(n_components=16)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents, columns = ['pc1', 'pc2','pc3','pc4','pc5','pc6','pc7','pc8','pc9','pc10','pc11','pc12','pc13','pc14','pc15','pc16'])
principalDf.head()
```

```python
pca.explained_variance_ratio_
```

```python
arr = pca.explained_variance_ratio_
sum_ = 0
for i in arr:
    sum_ += i
print(sum_)
```
